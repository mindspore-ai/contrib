```python
import matplotlib.pyplot as plt
import numpy as np
from positionalembedding2d import positionalencoding1d,positionalencoding2d
```

```python
# 1D Visualization
# create a PE for 100 data points, with model dimension 128.
pe = positionalencoding1d(128, 100)
plt.figure(figsize=(15, 5))
plt.plot(np.arange(100), pe[:,0:63:16].numpy())
plt.legend(["dim %d"%p for p in [4,5,6,7]])
```

```python
# 2D Visualization
# create a PE for 100*100 image, with model dimension 128.
pe2 = positionalencoding2d(128, 100, 100)
# show 4 different dimensions
plt.subplot(221)
plt.imshow(pe2[0,:,:].asnumpy())
plt.subplot(222)
plt.imshow(pe2[63,:,:].asnumpy())
plt.subplot(223)
plt.imshow(pe2[64,:,:].asnumpy())
plt.subplot(224)
plt.imshow(pe2[127,:,:].asnumpy())
```
